I saw some people using average daily returns and risk in their stock purchasing decisions. I wanted to learn about those so I will be using them this week. I read more about daily returns from https://www.wikihow.com/Calculate-Daily-Return-of-a-Stock and about risk/reward from https://www.investopedia.com/ask/answers/021915/how-standard-deviation-used-determine-risk.asp
I searched "tech company stocks" on Google and chose some that I knew. The csv files are from Yahoo Finance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mplt
aapl = pd.read_csv('week2/AAPL.csv')
amzn = pd.read_csv('week2/AMZN.csv')
fb = pd.read_csv('week2/FB.csv')
googl = pd.read_csv('week2/GOOGL.csv')
intc = pd.read_csv('week2/INTC.csv')
msft = pd.read_csv('week2/MSFT.csv')
orcl = pd.read_csv('week2/ORCL.csv')
pypl = pd.read_csv('week2/PYPL.csv')
tsla = pd.read_csv('week2/TSLA.csv')
```

Make the Date column values datetime objects

```python
aapl['Date'] = pd.to_datetime(aapl['Date'])
amzn['Date'] = pd.to_datetime(amzn['Date'])
fb['Date'] = pd.to_datetime(fb['Date'])
googl['Date'] = pd.to_datetime(googl['Date'])
intc['Date'] = pd.to_datetime(intc['Date'])
msft['Date'] = pd.to_datetime(msft['Date'])
orcl['Date'] = pd.to_datetime(orcl['Date'])
pypl['Date'] = pd.to_datetime(pypl['Date'])
tsla['Date'] = pd.to_datetime(tsla['Date'])
```

For the average daily returns, I read that adjusted closing price should be used because it is more accurate. I used the standard deviation of the closing price to help determine risk.

```python
avg_daily_returns = []
avg_daily_returns.append(np.mean(aapl['Adj Close']-aapl['Open']))
avg_daily_returns.append(np.mean(amzn['Adj Close']-amzn['Open']))
avg_daily_returns.append(np.mean(fb['Adj Close']-fb['Open']))
avg_daily_returns.append(np.mean(googl['Adj Close']-googl['Open']))
avg_daily_returns.append(np.mean(intc['Adj Close']-intc['Open']))
avg_daily_returns.append(np.mean(msft['Adj Close']-msft['Open']))
avg_daily_returns.append(np.mean(orcl['Adj Close']-orcl['Open']))
avg_daily_returns.append(np.mean(pypl['Adj Close']-pypl['Open']))
avg_daily_returns.append(np.mean(tsla['Adj Close']-tsla['Open']))
table = {'Name' : ['AAPL', 'AMZN', 'FB', 'GOOGL', 'INTC', 'MSFT', 'ORCL', 'PYPL', 'TSLA'],
        'Avg Returns' : avg_daily_returns,
         'Risk' : [aapl['Adj Close'].std(), amzn['Adj Close'].std(), fb['Adj Close'].std(), googl['Adj Close'].std(),
                  intc['Adj Close'].std(), msft['Adj Close'].std(), orcl['Adj Close'].std(), pypl['Adj Close'].std(),
                  tsla['Adj Close'].std()]
}
df = pd.DataFrame(data=table)
df
```

Finally, I plotted the average daily returns and the risk on a scatter plot to make it easier to visualize. 

```python
groups = df.groupby("Name")
for name, group in groups:
    mplt.plot(group["Risk"], group["Avg Returns"], marker="o", linestyle="", label=name)
mplt.legend()
mplt.xlabel('Risk')
mplt.ylabel('Avg Daily Returns')
```

And a graph of the the closing prices of the stocks

```python
ax = aapl.plot(x="Date", y="Adj Close", label='AAPL', figsize = (15, 10))
amzn.plot(ax=ax, x="Date", y="Adj Close", label='AMZN')
fb.plot(ax=ax, x="Date", y="Adj Close", label='FB')
googl.plot(ax=ax, x="Date", y="Adj Close", label='GOOGL')
intc.plot(ax=ax, x="Date", y="Adj Close", label='INTC')
msft.plot(ax=ax, x="Date", y="Adj Close", label='MSFT')
orcl.plot(ax=ax, x="Date", y="Adj Close", label='ORCL')
pypl.plot(ax=ax, x="Date", y="Adj Close", label='PYPL')
tsla.plot(ax=ax, x="Date", y="Adj Close", label='TSLA')
```
